```python
# KASUS 1

def input_matriks(baris, kolom, nama):
    print(f"\nMasukkan elemen matriks {nama} ({baris}x{kolom}):")
    matriks = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            while True:
                try:
                    val = float(input(f"Elemen [{i+1},{j+1}]: "))
                    row.append(val)
                    break
                except ValueError:
                    print("Input harus angka!")
        matriks.append(row)
    return matriks


def print_matriks(matriks):
    for row in matriks:
        print("\t".join(f"{elemen:.2f}" for elemen in row))


def penjumlahan_matriks(A, B):
    baris = len(A)
    kolom = len(A[0])
    C = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            row.append(A[i][j] + B[i][j])
        C.append(row)
    return C


def perkalian_matriks(A, B):
    m = len(A)
    p = len(A[0])
    n = len(B[0])
    C = []
    for i in range(m):
        row = []
        for j in range(n):
            total = 0
            for k in range(p):
                total += A[i][k] * B[k][j]
            row.append(total)
        C.append(row)
    return C


def transpose_matriks(A):
    return [[A[j][i] for j in range(len(A))] for i in range(len(A[0]))]


def main():
    print("\n=== KASUS 1: Matriks 2x2 ===")
    A = input_matriks(2, 2, "A")
    B = input_matriks(2, 2, "B")

    while True:
        print("\n===== MENU =====")
        print("1. Penjumlahan")
        print("2. Perkalian")
        print("3. Transpose")
        print("4. Keluar")
        pilih = input("Pilih operasi: ")

        if pilih == "1":
            print("\nHasil Penjumlahan:")
            print_matriks(penjumlahan_matriks(A, B))

        elif pilih == "2":
            print("\nHasil Perkalian:")
            print_matriks(perkalian_matriks(A, B))

        elif pilih == "3":
            print("\nTranspose Matriks A:")
            print_matriks(transpose_matriks(A))
            print("\nTranspose Matriks B:")
            print_matriks(transpose_matriks(B))

        elif pilih == "4":
            break

        else:
            print("Pilihan tidak tersedia!")

if __name__ == "__main__":
    main()
```
